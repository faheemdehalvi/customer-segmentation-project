# Data Cleaning

In this notebook, we will perform data cleaning steps to prepare the customer dataset for analysis. This includes handling missing values, removing cancelled orders, and ensuring the dataset is in a suitable format.

```python
import pandas as pd

# Load the dataset
data_path = '../data/raw/customers.csv'
customers = pd.read_csv(data_path)

# Display the first few rows of the dataset
customers.head()
```

## Handling Missing Values

We will check for missing values in the dataset and decide how to handle them.

```python
# Check for missing values
missing_values = customers.isnull().sum()
missing_values[missing_values > 0]
```

### Removing Cancelled Orders

We will remove any cancelled orders from the dataset.

```python
# Remove cancelled orders
customers = customers[customers['OrderStatus'] != 'Cancelled']

# Display the shape of the dataset after removal
customers.shape
```

## Final Dataset Preparation

We will perform any additional cleaning steps necessary to prepare the dataset for analysis.

```python
# Convert date columns to datetime format
customers['OrderDate'] = pd.to_datetime(customers['OrderDate'])

# Reset index
customers.reset_index(drop=True, inplace=True)

# Save the cleaned dataset
cleaned_data_path = '../data/processed/cleaned_customers.csv'
customers.to_csv(cleaned_data_path, index=False)

# Display the first few rows of the cleaned dataset
customers.head()
```
